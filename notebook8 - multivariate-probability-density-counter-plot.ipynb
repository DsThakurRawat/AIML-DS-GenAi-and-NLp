```python
""""
üî∑ What is a Contour Plot?
‚úÖ Definition:
A contour plot is a graphical technique for representing a 3D surface on a 2D plane, using contour lines. Each contour line represents points of equal value (like height, temperature, or probability density).

It is often used to visualize multivariate functions of the form:

ùëç
=
ùëì
(
ùëã
,
ùëå
)
Z=f(X,Y)
where X and Y are input variables and Z is the output.


""""
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import multivariate_normal

# Define grid
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Mean and covariance matrix
mu = [0, 0]
sigma = [[1, 0.5], [0.5, 1]]

# Compute the multivariate normal PDF
pos = np.dstack((X, Y))
rv = multivariate_normal(mu, sigma)
Z = rv.pdf(pos)

# Contour Plot
plt.figure(figsize=(6,5))
contour = plt.contour(X, Y, Z, levels=10, cmap='viridis')
plt.colorbar(contour)
plt.title("Contour Plot of 2D Gaussian Distribution")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
